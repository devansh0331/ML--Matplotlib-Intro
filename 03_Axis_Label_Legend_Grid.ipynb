```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]
max_temp = [50,51,52,48,47,49,48]
min_temp = [43,42,40,44,33,35,37]
avg_temp = [45,48,48,46,40,42,41]
```

```python
plt.plot(days , max_temp , label="max")
plt.plot(days , min_temp , label="min")
plt.plot(days , avg_temp , label="avg")

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather Report of 7 Days")

plt.legend(fontsize="small" , loc="best" , shadow="True")
plt.grid()

```
