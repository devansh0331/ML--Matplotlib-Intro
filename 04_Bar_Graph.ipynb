```python
import matplotlib.pyplot as plt
import numpy as np

company = ['GOOGL' , 'AMZN' , 'MSFT' , 'META']
rvn = [90 , 136 , 89 , 27]
profit = [20 , 50 , 10 , 8]
```

```python
xcomp = np.arange(len(company))
xcomp
```

```python
plt.xticks(xcomp , company)
plt.bar(xcomp , rvn , label='revenue')
plt.bar(xcomp , profit , label='profit')
plt.ylabel("Revenue (billions)")
plt.title("MNCs Annual Report")
plt.legend()
```
